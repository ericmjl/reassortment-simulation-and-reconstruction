# Introduction

In this notebook, I will run 500 simulations and plot the distribution of accuracy metrics from the reconstructed and shuffled graphs. This notebook will also serve as the "reference" notebook upon which other simulation runs will be performed.

## Step 1: Initialize viruses.

```python
from pathogen import Pathogen
from segment import Segment
from sequence import Sequence
from simulator import Simulator
from reconstructor import Reconstructor
from accuracy import *
```

```python
# Initialize sequences
sequence_PB2 = Sequence(length=1000) #this is representative of segment 1, PB2
sequence_H1 = Sequence(length=1000) #this is representative of segment 4, H1
sequence_H2 = Sequence(length=1000) #this is representative of segment 4, H2
sequence_H3 = Sequence(length=1000) #this is representattive of segment 4, H3
```

```python
# Initialize segments

from copy import deepcopy

sub_rate1 = 1E-2
sub_rate4 = 1E-1

## Here, we initialize two versions of segment_PB2, versions a and b, where b is slightly mutated from a.
segment_PB2a = Segment(segment_number=1, substitution_rate=sub_rate1, sequence=sequence_PB2)
segment_PB2b = segment_PB2a
segment_PB2b.mutate()

segment_H1 = Segment(segment_number=4, substitution_rate=sub_rate4, sequence=sequence_H1)
segment_H2 = Segment(segment_number=4, substitution_rate=sub_rate4, sequence=sequence_H2)
segment_H3 = Segment(segment_number=4, substitution_rate=sub_rate4, sequence=sequence_H3)
```

```python
# Add segments to seed viruses
virus_H1a = Pathogen(segments=[segment_PB2a, segment_H1], creation_time=0, progeny_size=(1.2, 0.5))
virus_H1b = Pathogen(segments=[segment_PB2b, segment_H1], creation_time=0, progeny_size=(1.2, 0.5))

virus_H2a = Pathogen(segments=[segment_PB2a, segment_H2], creation_time=0, progeny_size=(1.2, 0.5))
virus_H2b = Pathogen(segments=[segment_PB2b, segment_H2], creation_time=0, progeny_size=(1.2, 0.5))

virus_H3a = Pathogen(segments=[segment_PB2a, segment_H3], creation_time=0, progeny_size=(1.2, 0.5))
virus_H3b = Pathogen(segments=[segment_PB2b, segment_H3], creation_time=0, progeny_size=(1.2, 0.5))

viruses = [virus_H1a, virus_H1b, virus_H2a, virus_H2b, virus_H3a, virus_H3b]
viruses
```

```python
# Subclass the Simulator class for a specific implementation of the increment_timestep() function.
from random import random, choice, sample
class ViralSimulator(Simulator):
    
    def increment_timestep(self):
        cointoss = random()
        threshold = 0.75
        
        if cointoss <= threshold or (cointoss > threshold and len(self.pathogens) < 2):
            chosen_virus = choice(self.pathogens)
            new_virus = chosen_virus.replicate(current_time=self.current_time)
            
        if cointoss > threshold and len(self.pathogens) >= 2:
            [virus1, virus2] = sample(self.pathogens, 2)
            new_virus = virus1.reassort_with(virus2, current_time=self.current_time)
        
        self.pathogens.append(new_virus)
```

```python
# Define the number of runs
runs = 500
```

```python
for run_number in range(runs):
    # Randomly choose a number of viruses to seed the system.
    from random import randint, sample
    num_seeds = randint(1, (len(viruses)))
    seed_viruses = sample(viruses, num_seeds)

    # Initialize the viral simulator  with the seed viruses, and run.
    s = ViralSimulator()
    s.add_pathogens(seed_viruses)
    s.run(timesteps=50)
    
    # Write the Transmission Graphs and Sequences
    s.write_transmission_graph(outfile_name='Simulation Run %s' % run_number, folder_name='Simulation Network Pickles')
    s.write_sequences(outfile_name='Simulation Run %s' % run_number, folder_name='Simulation FASTA Files')
    
    # Perform Reconstruction with Full Sampling
    r = Reconstructor(segments=[1,4])
    r.read_fasta_file('Simulation FASTA Files/Simulation Run %s.fasta' % run_number)
    r.add_nodes_with_data()
    r.add_edges_with_weight()
    r.prune_graphs_by_weight()
    
    # Perform Reconstruction with Partial Sampling
    r2 = Reconstructor(segments=[1,4], sampling=True)
    r2.read_fasta_file('Simulation FASTA Files/Simulation Run %s.fasta' % run_number)
    r2.add_nodes_with_data()
    r2.add_edges_with_weight()
    r2.prune_graphs_by_weight()

    # Save the Ground Truth figure.
    fig = plt.figure()
    s.draw_transmission_graph(positions=True)
    fig = plt.gcf()
    fig.savefig('Simulation Network Figures/Simulation Run %s.pdf' % run_number)
    plt.close()

    # Save the Reconstruction figure.
    import networkx as nx
    fig = plt.figure()
    nx.draw(r.pruned_condensed_graph, pos=s.node_draw_positions)
    fig = plt.gcf()
    fig.savefig('Reconstruction Network Figures/Reconstruction Run %s.pdf' % run_number)
    plt.close()

    # Save the Null Model figure.
    fig = plt.figure()
    nx.draw(r.reassigned_source_graph, pos=s.node_draw_positions)
    fig = plt.gcf()
    fig.savefig('Null Model Network Figures/Null Model Run %s.pdf' % run_number)
    plt.close()
    
    # Save the Reconstruction figure under sampling
    fig = plt.figure()
    nx.draw(r2.pruned_condensed_graph, pos=s.node_draw_positions)
    fig = plt.gcf()
    fig.savefig('Sampled Reconstruction Network Figures/Reconstruction Run %s.pdf' % run_number)
    plt.close()
    
    # Save the Null Model figure under sampling.
    fig = plt.figure()
    nx.draw(r2.reassigned_source_graph, pos=s.node_draw_positions)
    fig = plt.gcf()
    fig.savefig('Sampled Null Model Network Figures/Null Model Run %s.pdf' % run_number)
    plt.close()

    # Append to the "alternate" (reconstruction) records
    alternate_fracts_accurate_reassortant_edges.append(fraction_accurate_reassortant_edges(s,r))
    alternate_fracts_inaccurate_reassortant_edges.append(fraction_inaccurate_reassortant_edges(s,r))
    alternate_fracts_accurate_edges.append(fraction_accurate_edges(s,r))
    alternate_fracts_inaccurate_edges.append(fraction_inaccurate_edges(s,r))
    alternate_fracts_accurate_reassortants.append(fraction_accurate_reassortants(s,r))
    alternate_fracts_inaccurate_reassortants.append(fraction_inaccurate_reassortants(s,r))
    alternate_fracts_path_accuracy.append(path_accuracy(s,r))
    alternate_fracts_reassortant_path_accuracy.append(reassortant_path_accuracy(s,r))

    # Append to the null model records
    null_fracts_accurate_reassortant_edges.append(fraction_accurate_reassortant_edges(s,r, reconstruction_type='reassigned_source'))
    null_fracts_inaccurate_reassortant_edges.append(fraction_inaccurate_reassortant_edges(s,r, reconstruction_type='reassigned_source'))
    null_fracts_accurate_edges.append(fraction_accurate_edges(s,r, reconstruction_type='reassigned_source'))
    null_fracts_inaccurate_edges.append(fraction_inaccurate_edges(s,r, reconstruction_type='reassigned_source'))
    null_fracts_accurate_reassortants.append(fraction_accurate_reassortants(s,r, reconstruction_type='reassigned_source'))
    null_fracts_inaccurate_reassortants.append(fraction_inaccurate_reassortants(s,r, reconstruction_type='reassigned_source'))
    null_fracts_path_accuracy.append(path_accuracy(s,r, reconstruction_type='reassigned_source'))
    null_fracts_reassortant_path_accuracy.append(reassortant_path_accuracy(s,r, reconstruction_type='reassigned_source'))
    
    # Append to the "alternate" (reconstruction) records under sampling
    sampled_alternate_fracts_accurate_reassortant_edges.append(fraction_accurate_reassortant_edges(s,r2))
    sampled_alternate_fracts_inaccurate_reassortant_edges.append(fraction_inaccurate_reassortant_edges(s,r2))
    sampled_alternate_fracts_accurate_edges.append(fraction_accurate_edges(s,r2))
    sampled_alternate_fracts_inaccurate_edges.append(fraction_inaccurate_edges(s,r2))
    sampled_alternate_fracts_accurate_reassortants.append(fraction_accurate_reassortants(s,r2))
    sampled_alternate_fracts_inaccurate_reassortants.append(fraction_inaccurate_reassortants(s,r2))
    sampled_alternate_fracts_path_accuracy.append(path_accuracy(s,r2))
    sampled_alternate_fracts_reassortant_path_accuracy.append(reassortant_path_accuracy(s,r2))
    
    # Append to the "null" model records under sampling
    sampled_null_fracts_accurate_reassortant_edges.append(fraction_accurate_reassortant_edges(s,r2, reconstruction_type='reassigned_source'))
    sampled_null_fracts_inaccurate_reassortant_edges.append(fraction_inaccurate_reassortant_edges(s,r2, reconstruction_type='reassigned_source'))
    sampled_null_fracts_accurate_edges.append(fraction_accurate_edges(s,r2, reconstruction_type='reassigned_source'))
    sampled_null_fracts_inaccurate_edges.append(fraction_inaccurate_edges(s,r2, reconstruction_type='reassigned_source'))
    sampled_null_fracts_accurate_reassortants.append(fraction_accurate_reassortants(s,r2, reconstruction_type='reassigned_source'))
    sampled_null_fracts_inaccurate_reassortants.append(fraction_inaccurate_reassortants(s,r2, reconstruction_type='reassigned_source'))
    sampled_null_fracts_path_accuracy.append(path_accuracy(s,r2, reconstruction_type='reassigned_source'))
    sampled_null_fracts_reassortant_path_accuracy.append(reassortant_path_accuracy(s,r2, reconstruction_type='reassigned_source'))
```
